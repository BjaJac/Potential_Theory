```python
import numpy as np
import pygimli as pg
from pygimli import meshtools as mt # import a module 
from pygimli.viewer import showMesh # import a function
```

```python
left = -1
right = 1
depth = 2
```

```python
world = mt.createWorld(start=[left, 1], end=[right, -1])
```

```python
circle = mt.createCircle(pos=(0, 0), radius=0.1, marker=2, nSegments=12)
```

```python
geometry = world + circle
pg.show(geometry);
```

```python
mesh = mt.createMesh(geometry, area=1, quality=34.6)
```
